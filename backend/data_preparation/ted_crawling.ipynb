```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time

#    
options = Options()
options.add_argument("headless")
options.add_argument("window-size=1920x1080")

#   
driver = webdriver.Chrome(options=options)

driver.get('https://www.youtube.com/c/TED/videos')

SCROLL_PAUSE_TIME = 2

# Get scroll height
last_height = driver.execute_script("return document.documentElement.scrollHeight")

while True:
    # Scroll down to the bottom
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")

    # Wait to load the page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# Once scrolling is done, retrieve the video titles

wait = WebDriverWait(driver, 30)
titles = wait.until(EC.presence_of_all_elements_located((By.ID, 'video-title')))
links = wait.until(EC.presence_of_all_elements_located((By.ID, 'video-title-link')))
times = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'div#thumbnail a#thumbnail div#overlays span#text')))


# video_titles = [video for video in titles]
video_urls = [link for link in links]
video_times = []
for time in times:
    if len(time.text.strip()) != 0:
        video_times.append(time)

video_data = [{"title": title.text, "link": link.get_attribute('href')[-11:], "text": time.text} for title, link, time in zip(titles, video_urls, video_times)]
for idx, data in enumerate(video_data, start=1):
    # print(f"{idx} {data['title']} ({data['link']}) - {data['text']}")
    if int(data['text'].split(":")[0]) >= 20:
        continue
    print(data['link'])

driver.quit()
```
